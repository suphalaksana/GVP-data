อ่านไฟล์ excel

```python
from google.colab import files
uploaded = files.upload() #โหลดไฟล์จากเครื่อง
```

```python
import io #โหลดข้อมูลเข้า Dataframe
#df = pd.read_csv(io.BytesIO(uploaded['ชื่อไฟล์ที่เราอัพโหลด.csv']))
df_order2 = pd.read_csv(io.BytesIO(uploaded['data upsell (1).csv']))
```

```python
df_order2
```

```python
order_data = df_order2[['Tel', 'Telesale', 'Product Name', 'Date']]
order_data
```

```python
duplicate_row_index = order_data.duplicated(subset=["Date", "Tel", "Telesale", "Product Name"], keep=False)
```

```python
duplicate_row_index
```

แสดงแถวที่ซ้ำ

```python
all_duplicate_rows = order_data[duplicate_row_index]
```

```python
all_duplicate_rows.to_excel("Duplicate_Rows.xlsx", index=True)
```
